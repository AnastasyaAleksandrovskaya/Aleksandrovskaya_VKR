```python
# Импортируем библиотеки
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Загружаем набор данных
data = pd.read_csv('databank2.csv', delimiter=';')
data.head()
```

```python
data.loc[data["sex"]=="Женский", "sex"] = 0
data.loc[data["sex"]=="Мужской", "sex"] = 1
data.head()
```

```python
data.loc[data["EducationStatus"]=="Незаконченное среднее образование", "EducationStatus"] = 0
data.loc[data["EducationStatus"]=="Среднее образование", "EducationStatus"] = 1
data.loc[data["EducationStatus"]=="Среднее специальное образование", "EducationStatus"] = 2
data.loc[data["EducationStatus"]=="Незаконченное высшее образование", "EducationStatus"] = 3
data.loc[data["EducationStatus"]=="Высшее образование", "EducationStatus"] = 4
data.loc[data["EducationStatus"]=="Несколько высших образований", "EducationStatus"] = 5
data.loc[data["EducationStatus"]=="Академическая степень", "EducationStatus"] = 6
data.head()
```

```python
data.loc[data["Residence"]=="село", "Residence"] = 0
data.loc[data["Residence"]=="город", "Residence"] = 1
data.head()
```

```python
data.loc[data["kolichestvo_rabotnikov_v_organizacii"]=="до 20", "kolichestvo_rabotnikov_v_organizacii"] = 0
data.loc[data["kolichestvo_rabotnikov_v_organizacii"]=="21-100", "kolichestvo_rabotnikov_v_organizacii"] = 1
data.loc[data["kolichestvo_rabotnikov_v_organizacii"]=="101-500", "kolichestvo_rabotnikov_v_organizacii"] = 2
data.loc[data["kolichestvo_rabotnikov_v_organizacii"]=="более 500", "kolichestvo_rabotnikov_v_organizacii"] = 3
data.head()
```

```python
data.loc[data["Employment"]=="Индивидуальный предприниматель", "Employment"] = 0
data.loc[data["Employment"]=="Сотрудник \ Рабочий \ Ассистент", "Employment"] = 1
data.loc[data["Employment"]=="Эксперт\Старший или Ведущий Специалист", "Employment"] = 2
data.loc[data["Employment"]=="Главный Специалист\Руководитель среднего звена", "Employment"] = 3
data.loc[data["Employment"]=="Руководитель высшего звена", "Employment"] = 4
data.head()
```

```python
data.pop('NaturalPersonID')
data.pop('RequestDate')
data.pop('ProductName')
data.pop('TypeOfWork')
data.pop('SignIP')
data.pop('LivingRegionName')
data.pop('IncomeDocumentKind')
data.pop('HaveSalaryCard')
data.pop('IsBankWorker')
data.pop('TypeOfWork.1')
data.pop('SignIP.1')
data.pop('otrasl_rabotodatelya')
data.pop('harakteristika_tekutschego_trudoustrojstva')
data.head()
```
